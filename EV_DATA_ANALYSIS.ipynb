## Removing Outlier

````python
def rem_out(col):
  sorted(col)
  Q1, Q3  = col.quantile([0.25, 0.75])
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)
  return lower_range, upper_range;
````

````python
lrbase, urbase  = rem_out(df['Base MSRP'])
df['Base MSRP'] = np.where(df['Base MSRP'] > urbase, urbase, df['Base MSRP'])
df['Base MSRP'] = np.where(df['Base MSRP'] < lrbase, lrbase, df['Base MSRP'])
````

````python
plt.figure(figsize=(10,5))
sns.boxplot(df['Base MSRP'])
plt.show()
````

````python
lrRange, urRange  = rem_out(df['Electric Range'])
df['Electric Range'] = np.where(df['Electric Range'] > urRange, urRange, df['Electric Range'])
df['Electric Range'] = np.where(df['Electric Range'] < lrRange, lrRange, df['Electric Range'])
````

````python
plt.figure(figsize=(10,5))
sns.boxplot(df['Electric Range'])
plt.show()
````
